### Cartesian coordinate system

```python
import matplotlib.pyplot as plt
import random as r
```

```python
x = 5
y = 7
plt.plot(x,y,'ro')
plt.show()
```

```python
""" This code creates a scatter plot of 5 random points within a 6x6 grid (from -3 to 3 on both axes), with each point displayed as a circle marker.
The axis('square') ensures the plot maintains equal scaling on both axes, and grid() adds gridlines to make it easier to read the coordinates.
Note that you'll need import random as r and import matplotlib.pyplot as plt at the beginning of your script for this code to work properly. """

# Loop to generate and plot 5 random points
for i in range(0,5):
    # Generate random x-coordinate between -3 and 3 (inclusive)
    x = r.randint(-3,3)
    # Generate random y-coordinate between -3 and 3 (inclusive)
    y = r.randint(-3,3)
    # Plot the point as a circle marker ('o') at coordinates (x,y)
    plt.plot(x,y,'o')

# Set the plot axes to have equal scaling (square aspect ratio)
plt.axis('square')
# Add a grid to the plot for better visualization
plt.grid()
# Display the completed plot
plt.show()
```

```python
"""This code creates a single arrow that starts at coordinates (2,1) and extends by a displacement vector of (5,6), meaning it ends at point (7,7).
The arrow has customized arrowhead dimensions, and the plot is constrained to show a 10x10 grid area.
The key thing to understand is that ax.arrow(x, y, dx, dy, ...) takes a starting point (x,y) and displacement
 values (dx,dy), not the ending coordinates directly."""

# Create a new axes object for the plot
# This is necessary to access matplotlib's arrow() method, which belongs to the axes object
# rather than the pyplot interface. While plt.plot() works directly, arrows require the axes.
ax = plt.axes()

# Draw an arrow starting at point (2,1) with displacement vector (5,6)
# The arrow will go from (2,1) to (7,7)
# head_width=0.2 sets the width of the arrowhead
# head_length=0.2 sets the length of the arrowhead
ax.arrow(2,1,5,6, head_width=0.2, head_length=0.2)

# Set the x-axis limits from 0 to 10
plt.xlim(0,10)
# Set the y-axis limits from 0 to 10
plt.ylim(0,10)

# Display the plot with the arrow
plt.show()
```
